```python
im_upload = SimpleNamespace(data = ["images/Leaf_Spots.jpg"])
im = PILImage.create(im_upload.data[-1])
im = im.to_thumb(256,256)
im
```

```python
import pathlib
plt = platform.system()
if plt == 'Linux': pathlib.WindowsPath = pathlib.PosixPath

learn = load_learner('model.pkl')
# learn = load_learner('C:/Users/marti/Desktop/DeepLearningforCoders/model.pkl')
```

```python
%time learn.predict(im)
```

```python
categories =  ["Downy Mildew","Leaf","Leaf Blights", "Leaf Rusts","Leaf Spots","Powdery Mildew"]

def classify_image(img):
    pred,idx,probs = learn.predict(img)
    return dict(zip(categories,map(float,probs)))
```

```python
classify_image(im)
```

```python
import gradio as gr
from gradio.components import Image, Label
intf = gr.Interface(title = "How is that leaf?",
                    description = "An image classifier that informs you which a desase a leaf has?",
                    fn=classify_image, 
                    inputs=Image(shape=(256,256)), 
                    outputs=Label(), 
                    examples=["images/Leaf_Spots.jpg","images/Powdery_Mildew.jpg","images/Rusts.jpg","images/Downy_Mildew.jpg","images/Healthy_Leaf.jpg","images/Leaf_Blights.jpg"],
                    live=True,
                    enable_queue=True
                    )
intf.launch()
```
